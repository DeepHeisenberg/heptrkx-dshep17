import and inline matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

make a random variable

```python
import numpy as np
```

```python
N=10000
x = np.random.normal(size=N)
```

make a one-D histogram

```python
plt.hist(x)
```

add a title to it

```python
plt.hist(x)
plt.title('A distribution')
```

change the color

```python
plt.hist(x, color='green')
plt.title('A distribution')
```

add a label

```python
plt.hist(x, label='gaussian', color='red', bins=20)
```

where is the label ?

add a legend

```python
plt.hist(x, color='pink',label='gaussian')
plt.title('A distribution')
plt.legend()
```

add another variable

```python
y = np.random.random(N)
```

```python
plt.hist(x, color='pink',label='gaussian')
plt.hist(y, color='red',label='flat')
plt.title('Distributions')
plt.legend()
```

change to the same binning

```python
bins = np.linspace(-4, 4, 20)
plt.hist(x, color='violet',label='gaussian', bins=bins)
plt.hist(y, color='blue',label='flat', bins=bins)
plt.title('Distributions')
plt.legend()
```

make it transparent

```python
bins = np.linspace(-4, 4, 20)
plt.hist(x, color='red',label='gaussian', bins=bins, alpha=0.5)
plt.hist(y, color='green',label='flat', bins=bins, alpha=0.5)
plt.title('Distributions')
plt.legend()
```

make a scatter plot

```python
n=50
plt.figure(figsize=(10,10))
plt.scatter(x[:n],y[:n], s=100, color ='pink', alpha=0.5, label='one direction')
plt.scatter(y[-n:],x[-n:], s=100, color ='blue', alpha=0.5, label='other direction')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```
